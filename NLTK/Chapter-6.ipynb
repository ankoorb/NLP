# Learning to Classify Text  

**Classification** is the task of choosing the correct **class label** for a given input. A classifier is called **supervised** if it is built based on training coropra containing the correct label for each input.

Supervised Classification: 
- Training:
    - Feature extractor is used to convert each input value to a feature set
    - Feature set + Labels -> Machine Learning Algorithm => Classifier Model
- Prediction: 
    - Feature extractor is used to convert unseen inputs to feature set
    - Feature set -> Classifier Model => Predicted Labels
    
    

- Decision Trees
- Naive Bayes Classifiers
- Maximum Entropy Classifiers

### Part-of-Speeh Tagging  

Decision Tree  

Classifier to work out which suffixes are most informative

```python
# Find most common suffixes
from nltk.corpus import brown
suffix_fdist = nltk.FreqDist()
for word in brown.words():
    word = word.lower()
    suffix_fdist[word[-1:]] += 1
    suffix_fdist[word[-2:]] += 1
    suffix_fdist[word[-3:]] += 1
```

```python
common_suffixes = [suffix for (suffix, count) in suffix_fdist.most_common(100)]
print common_suffixes
```

```python
# Feature extractor 
def pos_features(word):
    features = {}
    for suffix in common_suffixes:
        features['endswith({})'.format(suffix)] = word.lower().endswith(suffix)
    return features
```

```python
tagged_words = brown.tagged_words(categories = 'news')
featuresets = [(pos_features(n), g) for (n, g) in tagged_words]
```

```python
size = int(len(featuresets) * 0.1)
train_set, test_set = featuresets[size:], featuresets[:size]
```

```python
classifier = nltk.DecisionTreeClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

###Exploiting Context

- POS Classifier - Feature detector examines the context in which a word appears in order to determine which part of speech tag should be assigned. e.g. Tagging word 'fly', knowing that the previous word is 'a' - tag would be Noun, not a Verb. 

```python
def pos_features(sentence, i):
    features = {'suffix(1)': sentence[i][-1:],
                'suffix(2)': sentence[i][-2:],
                'suffix(3)': sentence[i][-3:]}
    
    if i == 0:
        features['prev-word'] = '<START>'
    else:
        features['prev-word'] = sentence[i-1]
    return features
```

```python
print brown.sents()[0]
print pos_features(brown.sents()[0], 8)
```

```python
tagged_sents = brown.tagged_sents(categories = 'news')
featuresets = []
for tagged_sent in tagged_sents:
    untagged_sent = nltk.tag.untag(tagged_sent)
    for i, (word, tag) in enumerate(tagged_sent):
        featuresets.append((pos_features(untagged_sent, i), tag))
```

```python
size = int(len(featuresets) * 0.1)
train_set, test_set = featuresets[size:], featuresets[:size]
classifier = nltk.DecisionTreeClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

### Sequence Classification


**Consecutive classification** or **Greedy Sequence classification** - Find the most likely class label for the first input, then use that answer to help find the best label for the next input

```python
def pos_features(sentence, i, history):
    features = {'suffix(1)': sentence[i][-1:],
                'suffix(2)': sentence[i][-2:],
                'suffix(3)': sentence[i][-3:]}
    
    if i == 0:
        features['prev-word'] = '<START>'
        features['prev-tag'] = '<START>'
    else:
        features['prev-word'] = sentence[i-1]
        # Each tag in history corresponds with a word in sentence
        features['prev-tag'] = history[i-1] 
    return features
```

```python
class ConsecutivePosTagger(nltk.TaggerI):
    
    def __init__(self, train_sents):
        train_set = []
        for tagged_sent in train_sents:
            untagged_sent = nltk.tag.untag(tagged_sent)
            history = []
            for i, (word, tag) in enumerate(tagged_sent):
                featureset = pos_features(untagged_sent, i, history)
                train_set.append((featureset, tag))
                history.append(tag)
        self.classifier = nltk.NaiveBayesClassifier.train(train_set)
        
    def tag(self, sentence):
        history = []
        for i, word in enumerate(sentence):
            featureset = pos_features(sentence, i, history)
            tag = self.classifier.classify(featureset)
            history.append(tag)
        return zip(sentence, history)
```

```python
tagged_sents = brown.tagged_sents(categories = 'news')
size = int(len(tagged_sents) * 0.1)
train_sents, test_sents = tagged_sents[size:], tagged_sents[:size]
tagger = ConsecutivePosTagger(train_sents)
print tagger.evaluate(test_sents)
```

Drawback: If a word is tagged as noun, and later it was found that the word should have been tagged as a verb - There is no way to go back and fix mistake. Use Brill tagger instead

###Sentence Segmentation  
Classification task for punctuation



```python
sents = nltk.corpus.treebank_raw.sents()
tokens = []
boundaries = set()
offset = 0
for sent in sents:
    tokens.extend(sent)
    offset += len(sent)
    boundaries.add(offset-1)
```

```python
def punct_features(tokens, i):
    return {'next-word-capitalized': tokens[i+1][0].isupper(),
            'prev-word': tokens[i-1].lower(),
            'punct': tokens[i],
            'prev-word-is-one-char': len(tokens[i-1]) == 1}
```

```python
featuresets = [(punct_features(tokens, i), (i in boundaries)) for i in range(1, len(tokens)-1) if tokens[i] in '.?!']
```

```python
size = int(len(featuresets) * 0.1)
train_set, test_set = featuresets[size:], featuresets[:size]
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

```python
def segment_sentences(words):
    start = 0
    sents = []
    for i, word in enumerate(words):
        if word in '.?!' and classifier.classify(punct_features(words, i)) == True:
            sents.append(words[start:i+1])
            start = i+1
        if start < len(words):
            sents.append(words[start:])
        return sents
```

### Evaluation

- Model Evaluation - Determine whether a classification model is accurately capturing a pattern
- Calculate a score for a model by comparing the labels that it generates for inputs in a test set or evaluation set with the correct labels for those inputs
- Accuracy - Measures the percentage of input in the test set that the classifier correctly labeled
- Precision - How many of the items that model retrieved were relevant, and Recall - How many of the relevant items that model retrieved

    - **True Positives (TP)** - Relevant items that model correctly identified as relevant
    - **True Negatives (TN)** - Irrelevant items that model correctly identified as irrelevant
    - **False Positives (FP)** - Irrelevant items that model incorrectly identfied as relevant
    - **False Negatives (FN)** - Relevant items that model incorrectly identfied as irrelevant

| | Condition Positive | Condition Negative |
|---|---|---|
| **Test Outcome Positive** | True Positive (relevant, retrieved) | False Positive (irrelevant, retrieved)|
| **Test Outcome Negative** | False Negative (relevant, not retrieved)| True Negative (irrelevant, not retrieved)|

- Confusion Matrix: **<font color = 'blue'>nltk.ConfusionMatrix(gold, test)</font>** - Matrix where each cell [i, j] indicate how often label j was predicted when the correct label was i. Diagonal entries indicate labels that were correctly predicted, and the off-diagonal entries indicate errors.

- Cross-Validation - Subdivide the original corpus into N subsets called folds. For each of these folds, train a model using all the data except the data in that fold, then test that model on the fold. If we get very similar scores for all N training sets, then we can be fairly confident that the score is accurate


###Decision Trees

- A simple flowchart that selects labels for input values. Decision nodes check feature values and leaf nodes assign labels. Flowcharts initial decision node is called root node, it contains a condition that checks one of the input value's features and selects a branch based on that feature's value. Following that branch we arrive at a new decision node with a new condition on the input value's features. We continue following the branch selected by each node's condition, until we arrive at a leaf node which provides a label for the input value.
- Decision stump - A decision tree with a single node that decides how to classify inputs based on a single feature, it contains one leaf for each possible feature value

**Entropy and Information Gain**

- Information Gain - Method for identifying the most informative feature for a decision stump. Measures how much more organized the input values become when we divide them up using a given feature. To measure how disorganized the original set of input values are, we calculate entropy of their labels, which will be high if the input values have highly varied labels, and low if many input values all have the same label. In particular, entropy is defined as the sum of the probability of each label times the log probability of that same label

- $H = -\sum_{l \in labels} P(l) * log_2(P(l))$ 
    - Labels with low frequency do not contribute to the entropy (since $P(l)$ is small)
    - Labels with high frequency do not contribute to the entropy (since $log_2(P(l)$ is small)

- Once we have calculated the entropy of the original set of input values' labels, we can determine how much more organized the labels become once we apply the decision stump. To do so, we calculate the entropy for each of the decision stump's leaves, and take the average of those leaf entropy values (weighted by the number of samples in each leaf). The information gain is then equal to the original entropy minus this new, reduced entropy. The higher the information gain, the better job the decision stump does of dividing the input values into coherent groups, so we can build decision trees by selecting the decision stumps with the highest information gain.
- Each branch of the decision tree splits the training data, the amount of training data available to train nodes lower in the tree can become quite small - Lower decision nodes may overfit the training set. Solution - Prune decision nodes that do not improve on a development test set
- Another problem - Decision trees require that features be checked in a specific order, this limits their ability to exploit features that ae relatively independent of one another

```python
# Entropy Calculation
import math
def entropy(labels):
    freqdist = nltk.FreqDist(labels)
    probs = [freqdist.freq(l) for l in freqdist]
    return -sum(p * math.log(p, 2) for p in probs)
```

### Naive Bayes Classifiers

- To choose a label for an input value, Naive Bayes Classifier begins by calculating the prior probability of each label (Determined by checking frequency of each label in the training set)  
- Assumption - All features are independent of one another given the label
- The contribution from each feature is then combined with this prior probability to arrive at a likelihood estimate for each label
- The label whose likelihood estimate is the highest is then assigned to the input value
- P(Feature|Label) = Count(Feature, Label) / Count(Label) - This approach can become problematic when a feature never occurs with a given label in the training set because P(Feature|Label) = 0, thus the input will be never assigned a label. Solution: Smoothing techniques for calculating P(Feature|Label). **<font color = 'blue'>nltk.probability</font>** module provides support for various smoothing techniques


### Maximum Entropy Classifiers

- Uses search techniques (iterative optimization) to find a set of parameters that will maximize the performance of the classifier. i.e. it find the set of parameters that will maximize the total likelihood of the training corpus
- Entropy - Measure of how "disorganized" a set of labels is. If a single label dominates then entropy is low, but if the labels are evenly distributed then entropy is high
- Chose the distribution with highest entropy

###Generative vs Conditional Classifiers

- Naive Bayes Classifier: Generative Classifier which builds a model that preditcs P(Input, Label), the joint probability of a (Input, Label) pair
- Maximum Entropy Classifier: Conditional Classifier which builds models that predict P(Label|Input) 
- The conditional probability P(Label|Input) can be calculated from the joint probability P(Input, Label) but not vice versa.
